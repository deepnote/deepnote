# Sales Data Analysis

A simple example demonstrating data analysis with pandas.

```python
import pandas as pd

# Sample sales data
data = {
    'product': ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard'],
    'units_sold': [150, 320, 95, 78, 210],
    'price': [999, 699, 449, 299, 79]
}

df = pd.DataFrame(data)
df['revenue'] = df['units_sold'] * df['price']
df
```

```python
# Calculate summary statistics
print(f"Total Revenue: ${df['revenue'].sum():,}")
print(f"Best Seller: {df.loc[df['units_sold'].idxmax(), 'product']}")
print(f"Highest Revenue: {df.loc[df['revenue'].idxmax(), 'product']}")
```
