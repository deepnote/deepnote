```python
class Table:
    def __init__(self, data):
        self.data = data
    def _repr_html_(self):
        rows = "".join(f"<tr><td>{k}</td><td>{v}</td></tr>" for k, v in self.data.items())
        return f"<table>{rows}</table>"
    def __repr__(self):
        return str(self.data)
```

```python
tables = []
for i in range(20):
    t = Table({f"key_{j}": f"val_{i}_{j}" for j in range(10)})
    display(t)
    tables.append(t)
```
