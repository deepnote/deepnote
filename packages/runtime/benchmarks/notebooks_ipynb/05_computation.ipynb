```python
data = list(range(10000))
```

```python
squared = [x**2 for x in data]
```

```python
total = sum(squared)
mean = total / len(squared)
```

```python
variance = sum((x - mean)**2 for x in squared) / len(squared)
std_dev = variance ** 0.5
```

```python
result = {
    "count": len(data),
    "sum": total,
    "mean": mean,
    "std_dev": std_dev,
}
```
