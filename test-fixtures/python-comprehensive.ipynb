```python
def calculate_sum(a, b):
    """Calculate the sum of two numbers.
    
    Args:
        a: First number
        b: Second number
    
    Returns:
        The sum of a and b
    """
    # Add the numbers together
    result = a + b
    return result
```

```python
class DataProcessor:
    """Process data with various transformations.
    
    This class provides methods for data cleaning and transformation.
    """
    
    def __init__(self, data):
        # Store the input data
        self.data = data
        self._processed = False
    
    def process(self):
        """Process the stored data."""
        # Mark as processed
        self._processed = True
        return self.data
```
